```python
y= df['class']
x = df.drop('class',axis=1)
```

```python
np.unique(y)
```

```python
y.value_counts()
```

```python
for columna in x.columns:
    print(f"{columna} tiene {x[columna].unique()}")
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()

for colomna in x.columns:
    x[columna] = encoder.fit_transform(x[[columna]])
```

```python
x.head()
```

```python
y = y.map(lambda x:1 if  x=='p' else 0)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

print(f" xTest : {x_test.shape[0]},{x_test.shape[1]}")
print(f" xTrain : {x_train.shape[0]},{x_train.shape[1]}")
print(f" yTest : {y_test.shape}")
print(f" yTrain : {y_train.shape}")
```

```python
from sklearn.naive_bayes import GaussianNB, CategoricalNB, BernoulliNB, MultinomialNB, ComplementNB
```

```python
algoritmos = {
    'GAUSIANO':GaussianNB(), 
    'CATEGORICAL' :CategoricalNB(), 
    'BERNOULLI' :BernoulliNB(), 
    'MULTINOMIAL': MultinomialNB(), 
    'COMPLEMENT' : ComplementNB()
    }
```

```python
y_pred = {}

for nombre, algoritmo in algoritmos.items():
    print(f"Entrenando: {nombre}")
    
    algoritmo.fit(x_train, y_train)
    y_pred[nombre] = algoritmo.predict(x_test)
```

```python
from sklearn
```

```python
def evualuacion(y_true, y_pred, metricas):
    res = {}
    for nombre, funcion in metricas.items():
        res[nombre] = funcion(y_true, y_pred)
        return res
```

```python
res = {}
for nombre, resultado in y_pred.items():
    res[nombre] = evaluacion(y_test, resultado, metricas) 
```
